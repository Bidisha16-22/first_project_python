Students category based on gender distribution

```python
plt.figure(figsize=(5,6))
ax=sns.countplot(data=df,x="Gender") #countplot counts the elements automatically
ax.bar_label(ax.containers[0]) #it counts the actual numbers in bar diagram
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have analysed that the number of females in the data is more than the number of males.
```

students score based on parents education

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,6))
sns.heatmap(gb,annot=True)
plt.title("Relation between parents education and student's score")
plt.show()
```

```python
#from the above chart we have concluded that education of parents can make impact on their child's education
```

Students score based on their parents marital status

```python
#result analysis according to parent's marital status
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Exam_scores.csv")
df.head()
mt=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(mt)
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(mt,annot=True)
plt.title("Relation between parents marital status and students score")
plt.show()
```

```python
#from the above chart we have concluded that the parents marital status cannot make any impact on their child's education
```

```python
#graphical representation of MathScore
sns.boxplot(data=df,x="MathScore")
plt.title("Graphical representation of Math")
plt.show()
```

```python
#graphical representation of ReadingScore
sns.boxplot(data=df,x="ReadingScore")
plt.title("Graphical representation of Reading")
plt.show()
```

```python
#graphical representation of WritingScore
sns.boxplot(data=df,x="WritingScore")
plt.title("Graphical representation of Writing")
plt.show()
```

```python
#from the above charts we have concluded that Students are weak in math compared to reading and writing skills.
```

Marks distribution based on Ethnic group

```python
print(df["EthnicGroup"].unique())
```

```python
groupa=df.loc[(df['EthnicGroup']=="group A")].count()
groupb=df.loc[(df['EthnicGroup']=="group B")].count()
groupc=df.loc[(df['EthnicGroup']=="group C")].count()
groupd=df.loc[(df['EthnicGroup']=="group D")].count()
groupe=df.loc[(df['EthnicGroup']=="group E")].count()
li=["group A","group B","group C","group D","group E"]
mylist=[groupa["EthnicGroup"],groupb["EthnicGroup"],groupc["EthnicGroup"],groupd["EthnicGroup"],groupe["EthnicGroup"]]
plt.pie(mylist,labels=li,autopct="%1.2f%%")
plt.title("Distribution of Ethnic group")
plt.show()
```
